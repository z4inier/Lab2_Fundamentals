```python
target_name = "MANOMBAGA"
SEED_NUM = "3"
redacted_name = ""

for char in target_name:
    if char in "AEIOU":
        redacted_name += SEED_NUM
    else:
        redacted_name += char

print(f"Original Surname: {target_name}")
print(f"SEED_NUM Used: {SEED_NUM}")
print(f"Redacted Output: {redacted_name}")
```

```python
#Forensic Signature
ascii_val = ord("M")
SEED_NUM = 3
loops = 6
verification_hash = (ascii_val * SEED_NUM) - loops

print("ascii_val:", ascii_val)
print("SEED_NUM:", SEED_NUM)
print("loops:", loops)  
print("verification_hash:", verification_hash)
```

```python
#Secure Input Loop
correct_password = "OpenSesame"
attempts = 0

while True:
    user_input = input("Enter the password: ")
    attempts += 1
    
    if user_input == correct_password:
        print("Access Granted!")
        break
    else:
        print("Incorrect password. Try again.")

print(f"Total attempts: {attempts}")
```
